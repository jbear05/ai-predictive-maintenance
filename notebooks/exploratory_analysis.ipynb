```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
### Basic Graph
x = [1, 2, 3]
y = [4, 5, 6]
plt.plot(x, y)

plt.title("Basic Line Graph", fontsize=16, fontdict = {'fontweight' : 'bold', 'color' : 'blue'})

plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

plt.show()
```
