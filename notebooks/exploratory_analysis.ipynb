```python
import pandas as pd

data = [
    {'Product': 'Laptop', 'Category': 'Electronics', 'Price': 1200, 'Quantity': 10},
    {'Product': 'T-Shirt', 'Category': 'Apparel', 'Price': 25, 'Quantity': 50},
    {'Product': 'Smartphone', 'Category': 'Electronics', 'Price': 800, 'Quantity': 5},
    {'Product': 'Headphones', 'Category': 'Electronics', 'Price': 150, 'Quantity': 20},
    {'Product': 'Jeans', 'Category': 'Apparel', 'Price': 75, 'Quantity': 30}
]

df = pd.DataFrame(data)

value = df.loc[(df['Category'] == 'Electronics') & (df['Price'] > 500), 'Price'].sum()

print("Total Price of Electronics over $500:", value)
```
